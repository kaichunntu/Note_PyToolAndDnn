## Convolutional Neural Network
Use [Fashion-mnist dataset](https://medium.com/tensorflow/hello-deep-learning-fashion-mnist-with-keras-50fcff8cd74a) from tf.keras
  
Label Description

    0 T-shirt/top 
    1 Trouser 
    2 Pullover 
    3 Dress 
    4 Coat 
    5 Sandal 
    6 Shirt 
    7 Sneaker 
    8 Bag 
    9 Ankle boot

```python
import os
import pickle
os.environ["CUDA_VISIBLE_DEVICES"]="0"

import numpy as np
import tensorflow as tf

sess_opt = tf.ConfigProto(gpu_options = \
                            ## Max ratio of memory of GPU can be used
                            tf.GPUOptions(per_process_gpu_memory_fraction=0.95 ,
                                          ## Allow memory occupatation growing while model needs more
                                          allow_growth=True ) 
                         ,allow_soft_placement=True ## allow swap
                         ,device_count={'GPU': 1}) ## Counts of GPU can be used

from matplotlib import pyplot as plt
```

#### Download dataset

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

idx2label = [ 
    "T-shirt/top", 
    "Trouser" ,
    "Pullover", 
    "Dress" ,
    "Coat" ,
    "Sandal", 
    "Shirt" ,
    "Sneaker", 
    "Bag" ,
    "Ankle boot"
]

print("Shape of training image :" , x_train.shape)
print("Shape of test image     :" , x_test.shape)
```

```python
def plot_img(data,label,size=5,idx=True):
    if idx:
        idx = np.random.choice(data.shape[0] , size , replace=False)
    else:
        idx = np.arange(size)
    
    data = data.reshape(-1,28,28)
    assert size%5 == 0 and size!=0
    fig , axList_ = plt.subplots(size//5,5 , figsize=[12,size//2] ,squeeze=False)
    
    axList = []
    for r in axList_:
        for c in r:
            axList.append(c)
    for i,ax in zip(idx,axList):
        ax.imshow(data[i],cmap="gray")
        ax.set_title(idx2label[label[i]])
#         ax.set_xlabel(idx2label[label[i]])
        ax.axis("off")
    
    plt.show()
```

```python
plot_img(x_train,y_train,15)
```

#### Normalize pixel value
There are some mechanism for normalizing.

    1. Normalize to Gaussian(0,1)
    2. Normalize to [0,1]
    3. Normalize to [-1,1]

```python
print("Max value of image :" , x_train.max())
print("Min value of image :" , x_train.min())
```

```python
## Normalize to [-1,1]
def preprocessing(data):
    return (data-128)/128-1

x_train = preprocessing(x_train)
x_test = preprocessing(x_test)
```

```python
print("After normalizing")
print("Max value of image :" , x_train.max())
print("Min value of image :" , x_train.min())
```

```python
x_train = x_train.reshape(-1,28,28,1)
x_test = x_test.reshape(-1,28,28,1)
```

## Build model
Model structure is like a graph which describes data flow. So write down the structure first and then build model by data flow.
  
Here, we would use module of tensorflow to construct model. i.e. tf.layers


```python
## Construct graph
fashion_g = tf.Graph() ## get a graph class
with fashion_g.as_default() as g:
    ## Build tensor for input data
    with tf.name_scope("Input"):
        _xs = tf.placeholder(tf.float32,shape=[None,*(x_train.shape[1::])] , name="image")
        _ys = tf.placeholder(tf.int32,shape=[None] , name="label")
        
        ## build one_hot vector for class
        onehot_y = tf.one_hot(_ys,depth=10,dtype=tf.float32)
        
    with tf.name_scope("Layer"):
        cnn1 = tf.layers.conv2d(_xs,32,[3,3],strides=[1,1],activation=tf.nn.leaky_relu,
                                padding="same",name="Conv1")
        print(cnn1)
        cnn2 = tf.layers.conv2d(cnn1,64,[3,3],strides=[2,2],activation=tf.nn.leaky_relu,
                                padding="same",name="Conv2")
        print(cnn2)
        cnn3 = tf.layers.conv2d(cnn2,64,[3,3],strides=[1,1],activation=tf.nn.leaky_relu,
                                padding="same",name="Conv3")
        print(cnn3)
        pool1 = tf.layers.max_pooling2d(cnn3,[3,3],strides=[2,2],padding="valid",name="Pool1")
        print(pool1)
        cnn4 = tf.layers.conv2d(pool1,64,[3,3],strides=[1,1],activation=tf.nn.leaky_relu,
                                padding="valid",name="Conv4")
        print(cnn4)
        
        flat = tf.layers.flatten(cnn4)
        
        ## fully-connected layer
        fc1 = tf.layers.dense(flat,128,activation=tf.nn.leaky_relu,name="FC1")
        print(fc1)
        ## prediction
        ## Produce a logit for image
        ## Then use softmax to get prob distribution
        _logits = tf.layers.dense(flat,10,activation=None,name="Prediction")
        print("Logit: " , _logits)
        _pred = tf.nn.softmax(_logits,axis=-1)
        _pred_label = tf.argmax(_pred,axis=-1,output_type=tf.int32)
        
    with tf.name_scope("Loss"):
        ## Write down objective function
        ## While this is classification task , use categorical cross entropy
        ce = tf.nn.softmax_cross_entropy_with_logits_v2(labels=onehot_y,logits=_logits)
        _loss = tf.reduce_mean(ce)
    
    with tf.name_scope("Training"):
        opt = tf.train.AdamOptimizer(learning_rate=5e-4)
        _update = opt.minimize(_loss)
    
    with tf.name_scope("Accuracy"):
        acc_bool = tf.equal(x=_ys,y=_pred_label)
        count = tf.cast(acc_bool,dtype=tf.float32)
        _acc = tf.reduce_mean(count)
    
    ## Export graph and visiualize it by tensorboard
    tf.summary.FileWriter("fashion_log",fashion_g)
    _init = tf.global_variables_initializer()
    
    ## Get all variable tensor
    all_var = tf.trainable_variables()
    ## Saving variable. 
    ## Max_to_keep means max number of model can be stored at same folder
    saver = tf.train.Saver(var_list=all_var,max_to_keep=5)
        
```

```python
def calculate_var_count(varList):
    c = 0
    for v in varList:
        d = 1
        for dd in v.shape:
            d*=int(dd)
        c+=d
    print("Count of variables :" , c)
calculate_var_count(all_var)
```

## Training

```python
# Open a session that would put tensor into GPU memory.
# After opening session, we can calculate any tensor by "sess.run(tensor , feed_dict)"
sess = tf.Session(graph=fashion_g,config=sess_opt)

## Initialize all variable
sess.run(_init)
```

```python
def get_batch(idx):
    d = {
        _xs:x_train[idx],
        _ys:y_train[idx]
    }
    return d

def batcher(r_index,n_step,batch_size=128):
    start = 0
    for s in range(n_step):
        yield s,get_batch(r_index[start:start+batch_size])
        start += batch_size
        
def evaluate_batch(sess,loss,acc,xs,ys,size=1000):
    idx = np.random.choice(x_test.shape[0],size,replace=False)
    l , a = sess.run([loss,acc],feed_dict={
        xs:x_test[idx],
        ys:y_test[idx]
    })
    return l,a
```

```python
batch_size = 256
n_epochs = 10
n_step = x_train.shape[0]//batch_size
r_index = np.arange(x_train.shape[0])
l,a = evaluate_batch(sess,_loss,_acc,_xs,_ys)
print("{0:>3d}/{1:>3d}  Loss : {2:>5.3f} | Acc : {3:>.4f}".format(0,n_epochs,l,a))

for e in range(1,n_epochs+1):
    np.random.shuffle(r_index)
    data_iter = batcher(r_index,n_step,batch_size=batch_size)
    for s,batch_data in data_iter:
        _,l = sess.run([_update,_loss],feed_dict=batch_data)
        print("{0:>4d}  loss : {1:>5.3f}".format(s,l) , end="\r")
    l,a = evaluate_batch(sess,_loss,_acc,_xs,_ys)
    print("{0:>3d}/{1:>3d}  Loss : {2:>5.3f} | Acc : {3:>.4f}".format(e,n_epochs,l,a))
        
    
saver.save(sess,"model.ckpt")
```

```python
def testPlot(size=5):
    idx = np.random.choice(x_test.shape[0],size,replace=False)
    pred_label = sess.run(_pred_label,feed_dict={_xs:x_test[idx]})
    print("Pred label")
    plot_img(x_test[idx],pred_label,size,False)
    print("Ground Truth")
    plot_img(x_test[idx],y_test[idx],size,False)
```

```python
testPlot(10)
```
